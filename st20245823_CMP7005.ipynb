
# **Methods Used for Handling Missing Values**

To keep the dataset suitable for time‑series analysis, different imputation techniques were applied depending on how each variable behaves over time. Instead of relying on simple mean or median filling—which can distort temporal patterns—the methods were chosen to preserve natural trends and realistic pollutant behavior.

**Time‑based interpolation** was used for pollutants that change rapidly from hour to hour, while **monthly median values** were applied to variables with more stable long‑term patterns. For gases with related chemical behavior, imputation followed known **scientific relationships** to maintain consistency.

Overall, these methods help retain realistic pollutant dynamics and produce cleaner, more reliable data for further analysis.

```python
combined_data['Date'] = pd.to_datetime(combined_data['Date'], dayfirst=True)
df = combined_data.sort_values(["station", "Date"])

cols_to_interp = ["PM2.5", "PM10", "CO","TEMP", "PRES", "DEWP", "RAIN", "WSPM"]

def interpolate_city(group):
    group = group.set_index("Date") #temporary index creation
    for col in cols_to_interp:
        group[col] = group[col].interpolate(
            method="time",
            limit=3,          # fill small gaps
            limit_direction="both"
        )
    return group.reset_index()

df = df.groupby("station", group_keys=False).apply(interpolate_city)
print(df[cols_to_interp].isna().sum())
```

```python
# Sort by Date only
df = df.sort_values("Date")

def interpolate_date(df):
    df2 = df.set_index("Date")
    for col in cols_to_interp:
        df2[col] = df2[col].interpolate(
            method="time",
            limit_direction="both"
        )
    return df2.reset_index()

df = interpolate_date(df)
print(df[cols_to_interp].isna().sum())
```

```python
# Extract month from Date column
df['Month'] = df['Date'].dt.month

# Columns that can use monthly median (numeric only)
cols_to_fill = ["TEMP", "PRES", "DEWP", "RAIN", "WSPM","SO2","NO2","O3"]

# Monthly median imputation for numeric columns
for col in cols_to_fill:
    df[col] = df.groupby('Month')[col].transform(lambda x: x.fillna(x.median()))

# Monthly mode imputation for wind direction (categorical)
df['wd'] = df.groupby('Month')['wd'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.ffill())
)

# Check remaining missing values
print(df[["TEMP", "PRES", "DEWP", "RAIN", "WSPM","SO2","NO2","O3"]].isna().sum())
```

# **Inference: AQI Calculation using CPCB Standards**

---

**Air Quality Index (AQI)** has been calculated following the **Central Pollution Control Board (CPCB), India** guidelines. This method is widely used in India for public communication of air quality.

### Key Features of the Implementation:

- **Breakpoints**: Official CPCB PM2.5 and PM10 concentration breakpoints were used.
- **Rounding Rule**: As per CPCB guidelines, pollutant concentrations were **rounded to the nearest integer** before computing sub-indices.
- **Sub-Index Calculation**: Linear interpolation method applied within each breakpoint range.
- **Final AQI**: Taken as the **maximum** of the PM2.5 and PM10 sub-indices (dominant pollutant approach).
- **AQI Categories**: Good, Satisfactory, Moderate, Poor, Very Poor, and Severe.

This allows direct comparison of Beijing’s air quality with Indian AQI standards and provides a standardized way to interpret pollution levels.

**Reference:**
- CPCB National Air Quality Index (AQI) Report  
  https://cpcb.nic.in/displaypdf.php?id=bmF0aW9uYWwtYWlyLXF1YWxpdHktaW5kZXgvRklOQUwtUkVQT1JUX0FRSV8ucGRm

```python


# PCB Breakpoints
pm25_breakpoints = [
    (0,   30,   0,   50),
    (31,  60,  51,  100),
    (61,  90, 101,  200),
    (91, 120, 201,  300),
    (121,250, 301,  400),
    (251,500, 401,  500),
]

pm10_breakpoints = [
    (0,    50,   0,   50),
    (51,  100,  51,  100),
    (101, 250, 101,  200),
    (251, 350, 201,  300),
    (351, 430, 301,  400),
    (431,1000, 401,  500),
]


# Robust Sub-index Function
def sub_index(conc, bps):
    if pd.isna(conc):
        return None
    
    conc = float(conc)
    
    if conc < 0:
        conc = 0
    
    max_conc = bps[-1][1]   
    if conc > max_conc:
        return bps[-1][3]   
    
    for low, high, i_low, i_high in bps:
        if low <= conc <= high:
            return (i_high - i_low) * (conc - low) / (high - low) + i_low
    
    return None  


# Round the concentrations (CPCB Rule)
df['PM2.5_round'] = df['PM2.5'].round().astype("Int64")
df['PM10_round']  = df['PM10'].round().astype("Int64")


# Compute Sub-indices
df['PM2.5_sub'] = df['PM2.5_round'].apply(lambda x: sub_index(x, pm25_breakpoints))
df['PM10_sub']  = df['PM10_round'].apply(lambda x: sub_index(x, pm10_breakpoints))


# Final AQI = Max of sub-indices
df['AQI_recalc'] = df[['PM2.5_sub', 'PM10_sub']].max(axis=1).round().astype('Int64')


# 6. AQI Bucket
def aqi_bucket_cpcb(aqi):
    if pd.isna(aqi):
        return None
    a = float(aqi)
    if a <= 50:
        return "Good"
    elif a <= 100:
        return "Satisfactory"
    elif a <= 200:
        return "Moderate"
    elif a <= 300:
        return "Poor"
    elif a <= 400:
        return "Very Poor"
    else:
        return "Severe"

df['AQI_Bucket_recalc'] = df['AQI_recalc'].apply(aqi_bucket_cpcb)



display(df[['Date', 'station', 'PM2.5', 'PM10', 
            'PM2.5_round', 'PM10_round', 
            'PM2.5_sub', 'PM10_sub', 
            'AQI_recalc', 'AQI_Bucket_recalc']].head(10))
```

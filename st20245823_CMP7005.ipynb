```python
# importing the necessary libraries for data analysis
import pandas as pd
import numpy as np
import glob
```

```python
# Combine all city data into one big table
city_files = glob.glob("Dataset/*.csv")

all_cities_data = []


for file_name in city_files:
    
    city_df = pd.read_csv(file_name)
    all_cities_data.append(city_df)
    print(f"Loaded: {file_name}")
    
    
combined_data = pd.concat(all_cities_data, ignore_index=True)


#Save the combined data to a new CSV file
#combined_data.to_csv("all_cities_combined.csv", index=False)


print(f"SUCCESS: Combined {len(city_files)} city files into one table with {len(combined_data)} total rows")
```

**Description of dataset columns:**



Year – The calendar year when the air quality measurement was taken.

Month – The month of the measurement, useful for seasonal or monthly trend analysis.

Day – The day of the month when the reading was recorded.

Hour – The hour of the day (0–23), allowing analysis of hourly pollution patterns.

PM2.5 – Fine particulate matter smaller than 2.5 micrometers. These particles penetrate deep into the lungs and are linked to serious health risks.

PM10 – Particulate matter smaller than 10 micrometers, including dust, smoke, and coarse particles.

SO2 – Sulfur dioxide, a pollutant produced mainly from burning coal or oil; contributes to respiratory irritation and acid rain.

NO2 – Nitrogen dioxide, a toxic gas from combustion sources like vehicles and industry; contributes to smog and respiratory issues.

CO – Carbon monoxide, a colorless gas from incomplete combustion; harmful at high concentrations.

O3 – Ground-level ozone, formed by chemical reactions in sunlight; a major component of smog and harmful to lungs.

TEMP – Ambient air temperature (°C) at the time of measurement.

PRES – Atmospheric pressure (hPa), useful for understanding weather conditions influencing pollution.

DEWP – Dew point temperature (°C), indicating moisture content in the air.

RAIN – Rainfall amount, typically in millimeters; rainfall can reduce particulate pollution.

wd – Wind direction, indicating the compass direction from which the wind is blowing (e.g., NW, NNE).

WSPM – Wind speed (m/s), which affects pollutant dispersion.


### **Dataset Structure**

The dataset contains **140,256 rows and 18 columns**. It includes hourly air‑quality pollutant measurements, meteorological variables (TEMP, PRES, DEWP, RAIN, wind direction and speed), and station names that represent the monitoring locations. The date information is stored across **year, month, day, and hour** columns.

There are two main types of columns in the dataset:

-   **Categorical columns:** _station_, _wd_, and the combined _date_ fields when treated as a categorical timestamp (3 columns)
    
-   **Numerical columns:** All pollutant concentrations (PM2.5, PM10, SO2, NO2, CO, O3) and meteorological variables (TEMP, PRES, DEWP, RAIN, WSPM), along with the index-like _No_ column (13 columns)



### **Data Completeness**

The dataset is largely complete, with **no missing values** in the columns: **year, month, day, hour, station, No**

The pollutant and weather variables show varying levels of missing data:

-   **Least missing:** _DEWP (0.10%)_, _TEMP (0.10%)_, _RAIN (0.10%)_, _PRES (0.10%)_, _WSPM (0.08%)_, _wd (0.56%)_
    
-   **Moderate missing:** _PM10 (1.50%)_, _PM2.5 (2.18%)_, _SO2 (2.18%)_, _O3 (2.80%)_, _NO2 (3.06%)_
    
-   **Most missing:** **CO (6.17%)** — the variable with the highest proportion of missing values in the dataset


```python
#checking Missing values
def missing_values_table(df):
    mis_val = df.isnull().sum()

  
    mis_val_percent = 100 * mis_val / len(df)

    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})

    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False)

    return mis_val_table

missing_values = missing_values_table(combined_data)
display(missing_values.style.background_gradient(cmap='Greens'))
```
